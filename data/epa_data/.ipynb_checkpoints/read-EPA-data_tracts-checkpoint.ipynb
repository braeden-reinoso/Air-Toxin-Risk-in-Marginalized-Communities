# EPA DATA - Tracts at National Level

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#### DATA ACCESS AVAILABLE HERE: https://www.epa.gov/ejscreen/download-ejscreen-data #### 
#### RELEVANT DATASET DOWNLOAD FROM CSV file (National Level) + Tract Level ####

EPA_full = pd.read_csv('EJSCREEN_2023_Tracts_with_AS_CNMI_GU_VI.csv', encoding='latin-1', low_memory=False)
```

```python
EPA_full.info()
```

```python
EPA_full.head()
```

```python
col_list = [name for name in EPA_full.columns]
col_list
```

```python
select_col=col_list[1:6]+col_list[29:43]
select_col
```

```python
EPA_cut=EPA_full[select_col]
```

```python
EPA_full['STATE_NAME'].unique()
```

```python
EPA_mass = EPA_cut[EPA_cut['STATE_NAME']=='Massachusetts']
EPA_mass['CNTY_NAME'].unique()
```

```python
EPA_cal = EPA_cut[EPA_cut['STATE_NAME']=='California']
EPA_cal['CNTY_NAME'].unique()
```

```python
EPA_texas = EPA_cut[EPA_cut['STATE_NAME']=='Texas']
EPA_texas['CNTY_NAME'].unique()
```

```python
EPA_suffolk = EPA_mass[EPA_mass['CNTY_NAME']=='Suffolk']
EPA_LA = EPA_cal[EPA_cal['CNTY_NAME']=='Los Angeles']
EPA_harris = EPA_texas[EPA_texas['CNTY_NAME']=='Harris']
```
